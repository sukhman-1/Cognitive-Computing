Question 1

```python
import numpy as np
import matplotlib.pyplot as plt
M=float(input("Enter value M: "))
x=np.linspace(-10,10,400)
y1=M*x**2
y2=M*np.sin(x)
plt.plot(x,y1,'r--',label='$y=Mx^2$')
plt.plot(x,y2,'b:',label='$y=M\\sin(x)$')
plt.title('Mathematical Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

Question 2

```python
import pandas as pd
import seaborn as sns
data={'Subject':['Math','Physics','Chemistry','Biology','CS'],
      'Score':[85,78,92,88,95]}
df=pd.DataFrame(data)
plt.figure(figsize=(8,5))
ax=sns.barplot(x='Subject',y='Score',data=df,hue='Subject',palette='viridis',legend=False)
for p in ax.patches:ax.annotate(f'{p.get_height()}',(p.get_x()+0.3,p.get_height()+0.5))
plt.title('Subject Scores')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

Question 3

```python
roll_no=int(input("Enter roll number: "))
np.random.seed(roll_no)
data=np.random.randn(50)
fig,axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].plot(np.cumsum(data),color='green')
axs[0,0].set_title('Cumulative Sum')
axs[0,0].set_xlabel("Index")
axs[0,0].set_ylabel("Sum")
axs[0,1].scatter(range(50),data+np.random.normal(0,0.5,50),color='red')
axs[0,1].set_title('Noisy Scatter')
axs[0,1].set_xlabel("Index")
axs[0,1].set_ylabel("Value")
axs[0,0].grid(True)
axs[0,1].grid(True)
axs[1,0].grid(True)
axs[1,1].grid(True)
plt.tight_layout()
plt.show()
```

```python
url="https://raw.githubusercontent.com/AnjulaMehto/MCA/refs/heads/main/company_sales_data.csv"
df=pd.read_csv(url)
plt.figure(figsize=(8,4))
sns.lineplot(x='month_number',y='total_profit',data=df,marker='o')
plt.title('Monthly Total Profit')
plt.grid(True)
plt.show()
plt.figure(figsize=(8,4))
products=df.columns[1:7]
for product in products:sns.lineplot(x='month_number',y=product,data=df,marker='o',label=product)
plt.title('Product Sales Trends')
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(8,4))
df[products].sum().plot(kind='bar',color=sns.color_palette('Set2'))
plt.title('Total Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```
