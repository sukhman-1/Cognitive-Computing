Question 1

```python
import numpy as np
import matplotlib.pyplot as plt
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
sum_all = np.sum(gfg)
sum_rows = np.sum(gfg, axis=1)
sum_cols = np.sum(gfg, axis=0)
print(f"Sum of all elements: {sum_all}")
print(f"Row-wise sum: {sum_rows}")
print(f"Column-wise sum: {sum_cols}")
```

Question 2

```python
#a
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
sorted_indices = np.argsort(array)
smallest_4 = np.sort(array)[:4]
largest_5 = np.sort(array)[-5:]
print(f"Sorted array: {sorted_array}")
print(f"Indices of sorted array: {sorted_indices}")
print(f"4 smallest elements: {smallest_4}")
print(f"5 largest elements: {largest_5}")

#b
array_b = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer_elements = array_b[array_b == array_b.astype(int)]
float_elements = array_b[array_b != array_b.astype(int)]
print(f"Integer elements only: {integer_elements}")
print(f"Float elements only: {float_elements}")
```

Question 3

```python
#a
initials_ascii_sum = ord('A') + ord('B')
X = initials_ascii_sum
sales = np.array([X, X+50, X+100, X+150, X+200])
#b
tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * (1 + tax_rate)
#c
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
#d
weekly_sales = np.tile(sales, (3, 1)) * (1 + np.arange(3).reshape(-1, 1) * 0.02)
print(f"Sales data: {sales}")
print(f"Taxed sales data: {taxed_sales}")
print(f"Discounted sales data: {discounted_sales}")
print(f"Weekly sales data:\n{weekly_sales}")
```

Question 4

```python
x_values = np.linspace(-10, 10, 100)
y_square = x_values**2
y_sin = np.sin(x_values)
y_exp = np.exp(x_values)
y_log = np.log(np.abs(x_values) + 1)

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(x_values, y_square)
plt.title("Y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.subplot(2,2,2)
plt.plot(x_values, y_sin)
plt.title("Y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.subplot(2,2,3)
plt.plot(x_values, y_exp)
plt.title("Y = e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.subplot(2,2,4)
plt.plot(x_values, y_log)
plt.title("Y = log(|x| + 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.tight_layout()
plt.show()
```
