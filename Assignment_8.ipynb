Code to convert each specie to numeria value

```python
import pandas as pd
import io
df=pd.read_csv(io.BytesIO(uploaded["Iris (2).csv"])) #read the csv file
df.drop(columns=["Id"],inplace=True,errors="ignore") #drop unnecessary "id" column
#Covert each specie to a numeric value
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df["Species"]=label_encoder.fit_transform(df["Species"])
print(dict(enumerate(label_encoder.classes_)))
```

Code to split data

```python
from sklearn.model_selection import train_test_split
x=df.drop(columns=["Species"])
y=df["Species"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Code to train logistic regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=200)
model.fit(x_train,y_train)
```

Code to test prediction

```python
y_pred=model.predict(x_test)
compare=pd.DataFrame({"Actual":y_test.values,"Predicted":y_pred})
print(compare.head(10))
```

Code to test with new input

```python
sample=[[5.1,3.5,1.4,0.2]]
prediction=model.predict(sample)
print(f"Prediction:{label_encoder.classes_[prediction[0]]}")
```
