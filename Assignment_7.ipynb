Question 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
roll=int(input("Enter roll number: "))
np.random.seed(roll)
sales=np.random.randint(1000,5001,(12,4))
df=pd.DataFrame(sales,
                columns=['Electronics','Clothing','Home & Kitchen','Sports'],
                index=pd.date_range('2023-01-01',periods=12,freq='ME').strftime('%b'))
print("First 5 rows:\n",df.head())
print("\nSummary stats:\n",df.describe())
df['Total Sales']=df.sum(axis=1)
df['Growth Rate']=df['Total Sales'].pct_change()*100
if roll%2==0:df['Electronics']*=0.9
else:df['Clothing']*=0.85
plt.figure(figsize=(10,6))
df[['Electronics','Clothing','Home & Kitchen','Sports']].plot(marker='o')
plt.title('Monthly Sales Trends')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['Electronics','Clothing','Home & Kitchen','Sports']])
plt.title('Sales Distribution')
plt.ylabel('Units')
plt.show()
```

```python
arr=np.array([[1,-2,3],[-4,5,-6]])
print("Absolute:\n",np.abs(arr))
print("Flattened percentiles:",np.percentile(arr.flatten(),[25,50,75]))
print("Column percentiles:\n",np.percentile(arr,[25,50,75],axis=0))
print("Row percentiles:\n",np.percentile(arr,[25,50,75],axis=1))
print("Flattened stats:",arr.flatten().mean(),np.median(arr.flatten()),arr.flatten().std())
print("Column stats:\nMeans:",arr.mean(axis=0),"\nMedians:",np.median(arr,axis=0),"\nStd:",arr.std(axis=0))
print("Row stats:\nMeans:",arr.mean(axis=1),"\nMedians:",np.median(arr,axis=1),"\nStd:",arr.std(axis=1))
```

Question 4

```python
def swap_list(lst,pos1,pos2):
    temp=lst[pos1]
    lst[pos1]=lst[pos2]
    lst[pos2]=temp
    return lst
print(swap_list([1,2,3,4],1,3))
```

Question 5

```python
def swap_set(s,item1,item2):
    if item1 in s and item2 in s:
        l=list(s)
        i1,i2=l.index(item1),l.index(item2)
        l[i1],l[i2]=l[i2],l[i1]
        return set(l)
    return s
print(swap_set({11,22,33,44},22,44))
```
